# Unified Dataset with Feature Selection

Turn the silver unified_dataset into gold after reducing features.

```python
import pandas as pd
import joblib
import json
import os
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# File paths
DATA_PATH = "/dsa/groups/casestudycf25/team02/silver/unified_dataset.csv"
FEATURE_SELECTION_PATH = "/dsa/groups/casestudycf25/team02/silver/keep_cols.pkl"
OUTPUT_DIR = '/dsa/groups/casestudycf25/team02/gold/unified_dataset.csv'
```

```python
df = pd.read_csv(DATA_PATH)

df.head()

df.shape
```

```python
# Load models and reference data
feature_selection_cols = joblib.load(FEATURE_SELECTION_PATH)

print(feature_selection_cols)

print(f"Features kept: {len(feature_selection_cols)}")
```

```python
include_cols = [
   'target', 'npi', 'specialty', 'specialty_type', 'rfrg_prvdr_state_abrvtn', 'active_medicaid_provider_flag'
]

full_cols = include_cols + feature_selection_cols


gold_cols = [
    c for c in  df.columns if c in full_cols
]
```

```python
# Filter to columns that exist
df[gold_cols].shape
```
