## Attempt - 1 (XGBoost with Default K-best XGB)

Scikit-learns default k value is 10.

```python
# Grid 1: Tree structure with learning rate
param_grid_1 = [
    {
        "model__max_depth": [3, 5],
        "model__min_child_weight": [1, 3],
        "model__learning_rate": [0.05, 0.1],
        "model__n_estimators": [100, 200]
    }
]

# Grid 2: Sampling with regularization
param_grid_2 = [
    {
        "model__subsample": [0.6, 1.0],
        "model__colsample_bytree": [0.6, 1.0],
        "model__reg_alpha": [0, 0.1],
        "model__reg_lambda": [1, 5]
    }
]

# Grid 3: Imbalance handling
param_grid_3 = [
    {
        "model__scale_pos_weight": [50, 100, 200],
        "model__max_delta_step": [1, 5],
        "model__min_child_weight": [1, 5],
        "model__learning_rate": [0.01, 0.05],
        "model__n_estimators": [200, 300]
    }
]
```

The above parameters determine how well the model will perform. Based on the performance, these parameters will be adjusted.

```python
pipe = MLPipeline(model_name="xgboost", target_col="target")
```

```python
# param_grid_1
list = MLPipeline.list_undersample_methods()
for l in list:
    pipe.train(df, param_grid=param_grid_1, group_col="npi", exclude_cols=exclude_cols, undersample_method=l, feature_selection_method='k_best_xgb')
```

```python
# param_grid_2
list = MLPipeline.list_undersample_methods()
for l in list:
    pipe.train(df, param_grid=param_grid_2, group_col="npi", exclude_cols=exclude_cols, undersample_method=l, feature_selection_method='k_best_xgb')
```

```python
# param_grid_3
list = MLPipeline.list_undersample_methods()
for l in list:
    pipe.train(df, param_grid=param_grid_3, group_col="npi", exclude_cols=exclude_cols, undersample_method=l, feature_selection_method='k_best_xgb')
```

## Obervations:
With the K value begin set as 10, with a score of 0.2399, these parameters performed the best: 
1) model__learning_rate: 0.01
2) model__max_delta_step: 5
3) model__min_child_weight: 1
4) model__n_estimators: 300
5) model__scale_pos_weight: 50
6) Undersampling : RandomUnderSampler

When using cleanse clustering, most of the performance also saw a significant dip. Further testing will have to be conducted to see how it performs. 

Maybe adjusting the number of K values for the feature selection copuld yield us some better results.
