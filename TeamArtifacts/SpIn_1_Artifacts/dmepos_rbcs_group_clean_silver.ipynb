## RBCS_Grouping

The purpose of this notebook is to create a dataset containing every unique combination of RBCS levels, ID and description, and HCPCS code and description.

This dataset is intended for inclusion in the final database.

End Result:

Saved the cleaned dataset to a shared team directory: /dsa/groups/casestudycf25/team02/silver/dmepos_rbcs_group_clean.csv

```python
# Loading the required Libraries and Dataset
from pathlib import Path
import pandas as pd
df=pd.read_csv("/dsa/groups/casestudycf25/team02/silver/dmepos_rfrhpr_ser.csv",index=False)
```

```python
#Column names
df.columns
```

```python
# Selecting only 'RBCS_Lvl','RBCS_Id','RBCS_Desc','HCPCS_CD', 'HCPCS_Desc'
order = ['RBCS_Lvl','RBCS_Id','RBCS_Desc','HCPCS_CD', 'HCPCS_Desc']
rbcs_df = df[order]
```

```python
# First few rows
rbcs_df.head()
```

```python
# Shape
rbcs_df.shape
```

```python
# Unique combinations of RBCS level, ID, description, and HCPCS code and description
unique_grouping = rbcs_df[['RBCS_Lvl','RBCS_Id','RBCS_Desc','HCPCS_CD','HCPCS_Desc']].drop_duplicates()
```

```python
# First few rows
unique_grouping.head()
```

```python
# Shape
unique_grouping.shape
```

```python
# Saving file
unique_grouping.to_csv("/dsa/groups/casestudycf25/team02/silver/dmepos_rbcs_group_clean.csv", index=False)
```
